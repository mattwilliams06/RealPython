````python
import pandas as pd
import numpy as np
import urllib
import requests
import os
````

````python
def download_data(url, fname, ext):
    path = os.getcwd()
    target_path = f'{fname}.{ext}'
    response = requests.get(url)
    print(f'Response: {response.raise_for_status()}')
    with open(target_path, 'wb') as file:
        file.write(response.content)
    print(f'Download of {target_path} completed to {path}')
````

````python
precip_data = 'https://raw.githubusercontent.com/realpython/materials/master/introduction-combining-data-pandas-merge-join-and-concat/climate_precip.csv'
````

````python
download_data(precip_data, 'climate-precip', 'csv')
````

# Inner Join

Here, we'll use the default merge argument ```inner```, which will cause us to lose any rows that aren't matched between the datasets (unmatched key column rows). 

````python
precip_one_station = climate_precip[climate_precip['STATION']=='GHCND:USC00045721']
````

````python
precip_one_station.head()
````

````python
inner_merged = pd.merge(precip_one_station, climate_temp)
````

````python
inner_merged.head()
````

````python
inner_merged.shape
````

````python
precip_one_station.shape
````

````python
climate_temp[climate_temp['STATION'] == 'GHCND:USC00045721']
````

The key column is matched for all of the rows for Mitchell Caverns between the datasets, so we didn't lose any rows during the merge. We gained all of the columns from the temperature dataset.

# Outer Join

````python
outer_merged = pd.merge(precip_one_station, climate_temp, how='outer', on=['STATION', 'DATE'])
````

````python
outer_merged.head()
````

````python
outer_merged.shape
````

Here, the number of rows is the same as the original number for the larger dataset, because no rows were lost. we picked up the additional columns.

# Left Merge

In a left merge, the resulting dataframe will contain all of the rows in the left dataset, while discarding all of the unmatched rows from the right dataset.

````python
left_merged = pd.merge(climate_temp, precip_one_station, how='left', on=['STATION', 'DATE'])
````

````python
left_merged.shape
````

````python
left_merged_reversed = pd.merge(precip_one_station, climate_temp, how='left', on=['STATION', 'DATE'])
````

````python
left_merged_reversed.shape
````

# Right Merge

This is identical to the left merge, just with the operation reversed based off of the order of the arguments.

# Pandas ```join()```

````python
precip_one_station.join(climate_temp, lsuffix='_left', rsuffix='_right')
````

Upon inspection, this looks like a left join. However, columns that are common to both are not discarded, and we have appended them with suffixes to provide a distinction. 

````python
climate_temp.join(precip_one_station, lsuffix='_left', rsuffix='_right')
````

````python
inner_merged_total = pd.merge(climate_temp, climate_precip, on=['STATION', 'DATE'])
inner_merged_total.head()
````

````python
inner_joined_total = climate_temp.join(climate_precip.set_index(['STATION', 'DATE']),
                                      lsuffix='_x', rsuffix='_y', on=['STATION', 'DATE'])
inner_joined_total.head()
````

# Concatenation

Concatenating two DataFrames stitches them together along an axis.

````python
double_precip = pd.concat([precip_one_station, precip_one_station])
````

````python
double_precip.head()
````

````python
double_precip.shape
````

````python
reindexed = pd.concat([precip_one_station, precip_one_station], ignore_index=True)
````

````python
reindexed.head()
````

````python
outer_joined = pd.concat([climate_precip, climate_temp])
````

````python
outer_joined.head()
````

The ```NaN``` values appear because we care concatenating along axis 0, but the column indices do not match. This is an outer join. To drop the columns that do not have matching indices, we need an inner join.

````python
inner_joined = pd.concat([climate_precip, climate_temp], join='inner')
````

````python
inner_joined.head()
````

````python
inner_joined_cols = pd.concat([climate_precip, climate_temp], axis=1, join='inner')
````
