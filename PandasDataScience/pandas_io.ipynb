# SQL Files

````python
from sqlalchemy import create_engine
````

````python
engine = create_engine('sqlite:///data.db', echo=False)
````

````python
dtypes = {'POP': 'float64', 'AREA': 'float64', 'GDP': 'float64', 'IND_DAY': 'datetime64'}
````

````python
df = pd.DataFrame(data=data).T.astype(dtype=dtypes)
````

````python
df.dtypes
````

````python
df.to_sql('data.db', con=engine, index_label='ID')
````

````python
df = pd.read_sql('data.db', con=engine, index_col='ID')
````

````python
df
````

Creating the dataframe from a SQL database inserts an additional row after the header row. This is corrected below.

````python
df.index.name = None
````

````python
df
````

## Chunk Size

Using chunks creates an iterable (like a generator) that processes only a portion of the data at a time.

````python
data_chunk = pd.read_csv('countries.csv', index_col=0, chunksize=8)
````

````python
type(data_chunk)
````

````python
hasattr(data_chunk, '__iter__')
````

````python
hasattr(data_chunk, '__next__')
````

````python
for df_chunk in pd.read_csv('countries.csv', index_col=0, chunksize=8):
    print(df_chunk, end='\n\n')
    print(f'memory: {df_chunk.memory_usage().sum()} bytes', end='\n\n\n')
````
